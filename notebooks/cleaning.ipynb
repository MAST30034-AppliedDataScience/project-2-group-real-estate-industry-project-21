```python
# Imports
import pandas as pd
import json
import re
```

```python
# Load the data from the JSON file
with open('/Users/jennymai/Desktop/data_sci/project2/data/landing/domain_data.json', 'r') as f:
    data = json.load(f)

# Convert the JSON to a pandas DataFrame
df = pd.DataFrame.from_dict(data, orient='index')

df
```

```python
df.dtypes
```

```python
# 1. Split 'rooms' into separate 'Beds' and 'Baths' columns
def extract_beds_baths(rooms):
    beds, baths = None, None
    for room in rooms:
        if 'Bed' in room:
            beds = int(re.search(r'\d+', room).group())
        if 'Bath' in room:
            baths = int(re.search(r'\d+', room).group())
    return beds, baths

df['Beds'], df['Baths'] = zip(*df['rooms'].apply(extract_beds_baths))
df
```

```python
# 2. Extract numeric 'Parking' information
def extract_parking(parking_list):
    for parking in parking_list:
        match = re.search(r'\d+', parking)
        if match:
            return int(match.group())  # Extract the number if present
    return 0  # Default to 0 if no valid parking info is found

# Apply this function to the 'parking' column
df['Parking'] = df['parking'].apply(extract_parking)
df
```

```python
# 3. Extract numeric cost from 'cost_text'
def extract_numeric_cost(cost_text):
    match = re.search(r'\d+', cost_text.replace(',', ''))
    return int(match.group()) if match else None

df['Cost'] = df['cost_text'].apply(extract_numeric_cost)
df
```

```python
# Function to extract address components, handling possible variations
def extract_address_components(name):
    # Split the name based on commas to separate street and suburb/state/postcode
    parts = name.split(',')
    
    if len(parts) == 2:
        # Street address and suburb, state, postcode
        street = parts[0].strip()  # Street address before the first comma
        suburb_state_postcode = parts[1].strip()  # Suburb, state, postcode after the comma
        
        # Use regex to extract state and postcode (e.g., "VIC 3000")
        state_postcode_match = re.search(r'([A-Z]+)\s*(\d+)', suburb_state_postcode)
        if state_postcode_match:
            state = state_postcode_match.group(1)
            postcode = state_postcode_match.group(2)
        else:
            state, postcode = None, None
        
        # Extract suburb (the part before state/postcode)
        suburb = re.sub(r'[A-Z]+\s*\d+', '', suburb_state_postcode).strip()
    else:
        # Handle unexpected cases by returning None for each component
        street, suburb, state, postcode = None, None, None, None
    
    return street, suburb, state, postcode

# Apply the function to the 'name' column
df['Street Address'], df['Suburb'], df['State'], df['Postal Code'] = zip(*df['name'].apply(extract_address_components))
df
```
